```python
import os
import pandas as pd
import json
import gzip

downloads_df = None

for test in os.listdir('./data/raw'):
  if not os.path.exists(f'./data/raw/{test}/ndt7'):
    print(f'Skipping {test}, no ndt7 directory found')
    continue
  filename = None
  compressed = False
  for ndt7_file in os.listdir(f'./data/raw/{test}/ndt7'):
    if ndt7_file.startswith('ndt7-download') and ndt7_file.endswith('.json') or ndt7_file.endswith('.json.gz'):
      compressed = ndt7_file.endswith('.json.gz')
      filename = ndt7_file
      break
  if filename is None:
    print(f'No ndt7-download file found in {test}')
    continue
  with (gzip.open(f'./data/raw/{test}/ndt7/{filename}', 'rt') if compressed else open(f'./data/raw/{test}/ndt7/{filename}', 'r')) as f:
    data = json.load(f)
    if not data.get('Download') or not data.get('Download').get('ServerMeasurements'):
      print(data.get('Download'))
      print(f'No download measurements found in {test}')
      continue
    measurements = data['Download']['ServerMeasurements']
    last_measurement = measurements[-1]
    data['Mbps'] = last_measurement['TCPInfo']['BytesAcked'] / last_measurement['TCPInfo']['ElapsedTime'] * 8
    print(test, data['Mbps'])
```
